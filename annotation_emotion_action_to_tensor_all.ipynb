```python
from glob import glob
import json
import shutil
import tqdm
import cv2

action_list = ['BODYLOWER','BODYSCRATCH','BODYSHAKE','FEETUP','FOOTUP','HEADING','LYING','MOUNTING','SIT','TAILING','TAILLOW','WALKRUN']
for action in action_list:
    folder_path = f"/home/work/pose/image/{action}/"
    json_path = f"/home/work/pose/label/{action}/"
    folder_temp = glob(f"/home/work/pose/image/{action}/*")
    folder_list = list(map(lambda x : x.split('/')[-1], folder_temp))
    json_list = list(map(lambda x : x +'.json',folder_list))

    for i in tqdm.tqdm(range(len(folder_list))):
        frame_list = glob(folder_path+folder_list[i]+'/*')
        with open(json_path+json_list[i], "r") as st_json:
            json_data = json.load(st_json)
        for frame in frame_list:
            number = frame.split('/')[-1].split('_')[1]
            img = cv2.imread(frame)
            for j in json_data['annotations']:
                if j['frame_number'] == int(number):
                    shutil.copyfile(frame,'images/'+'train/'+frame.split('/')[-2]+'_'+frame.split('/')[-1])
                    #shutil.copyfile(frame,'images/','label/'+frame.split('/')[-2]+'_'+frame.split('/')[-1])
                    f = open('labels/'+frame.split('/')[-2]+'_'+frame.split('/')[-1].split('.')[0]+'.txt', 'w')
                    x = j['bounding_box']['x']
                    y = j['bounding_box']['y']
                    w = j['bounding_box']['width']
                    h = j['bounding_box']['height']

                    iw = img.shape[1]
                    ih = img.shape[0]

                    xx = (x+(w/2)) /iw
                    yy = (y+(h/2)) /ih
                    ww = w / iw
                    hh = h / ih
                    data = f'0 {xx} {yy} {ww} {hh}'

                    key_idx = [str(i+1) for i in range(15)]

                    for k in key_idx:
                        if j['keypoints'][k] is not None:
                            kx = j['keypoints'][k]['x'] / iw
                            ky = j['keypoints'][k]['y'] / ih
                            data += f' {kx} {ky}'
                        else:
                            data += ' 0 0'
                    f.write(data)
                    f.close()
                    break
```
