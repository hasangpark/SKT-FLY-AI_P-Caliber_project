```python
import os
import cv2
import random
def split_videos_random_frames(videos_folder, output_folder, frame_interval=6):
    # 동영상 폴더 내의 동영상 파일 목록 가져오기
    video_files = [f for f in os.listdir(videos_folder) if f.endswith('.mp4')]
    
    for video_file in video_files:
        video_path = os.path.join(videos_folder, video_file)
        output_subfolder = os.path.join(output_folder, os.path.splitext(video_file)[0])
        
        # 출력 서브폴더가 없으면 생성
        os.makedirs(output_subfolder, exist_ok=True)
        
        # 동영상 파일 열기
        cap = cv2.VideoCapture(video_path)
        frame_count = 0
        selected_frames = set()

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            # num_frames 개수만큼 랜덤한 프레임 선택
            if frame_count % frame_interval == 0:
                selected_frames.add(frame_count)
                output_path = os.path.join(output_subfolder, f"frame_{frame_count:04d}.jpg")
                cv2.imwrite(output_path, frame)

            frame_count += 1

        cap.release()

if __name__ == "__main__":
    input_videos_folder = "eating"    # 입력 동영상 폴더 경로
    output_frames_folder = "frame_result"  # 프레임 저장 폴더 경로

    num_frames_to_extract = 10  # 추출할 프레임 개수
    frame_interval = 6          # 프레임 추출 간격

    split_videos_random_frames(input_videos_folder, output_frames_folder,frame_interval)
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import json

class SimpleCNN3D(nn.Module):
    def __init__(self, num_classes):
        super(SimpleCNN3D, self).__init__()
        self.conv1 = nn.Conv3d(3, 16, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool3d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv3d(16, 32, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool3d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(32 * 4 * 4 * 4, 128)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.pool1(torch.relu(self.conv1(x)))
        x = self.pool2(torch.relu(self.conv2(x)))
        x = x.view(-1, 32 * 4 * 4 * 4)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```
